```python
import matplotlib.pyplot as plt

def plot_terrain(a, b, c):
    total_cells = a * b
    positions_x = []
    positions_y = []
    marker_size = c * 5000  # Taille du marqueur en fonction de la distance entre les cellules
    
    for i in range(a):
        for j in range(b):
            positions_x.append(j * c)
            positions_y.append(i * c)
    
    plt.figure(figsize=(5, 3.5))
    plt.scatter(positions_x, positions_y, s=marker_size, marker='s')  # Utilisation de marker_size
    plt.title(f"Repr√©sentation graphique de {total_cells} cellules")
    plt.xlabel('Position en X')
    plt.ylabel('Position en Y')
    plt.grid(True)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.show()


X = 11
a = 3
b = 4
c = 1

print(f"X cellules : {X}, {a} lignes, {b} colonnes, Distance entre cellules : {c}")

plot_terrain(a, b, c)
```
